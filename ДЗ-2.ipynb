```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
page = requests.get("https://www.theguardian.com/lifeandstyle/health-and-wellbeing")
page.text
soup = BeautifulSoup(page.text, 'lxml')
```

```python
def f(l):
    n = []
    for i in l:
        if i not in n:
            n.append(i)
    return n
```

ссылка на текст

```python
hrefs = []
titles = soup.find_all('li', class_="fc-slice__item l-row__item l-row__item--span-1 u-faux-block-link")
for title in titles:
    hrefs.append(title.find('a').get('href'))
hrefs = f(hrefs)
```

заголовки публикаций

```python
titles = []
for title1 in hrefs:
    html_title = requests.get(title1)
    html_title.text
    html_title = BeautifulSoup(html_title.text, 'lxml')
    for title_itself in html_title:
        title_itself = html_title.find('h1').text
        if title_itself not in texts:
            titles.append(title_itself)
titles = f(titles)                   
```

текст публикации

```python
texts = []
for text1 in hrefs:
    html_title = requests.get(text1)
    html_title.text
    html_title = BeautifulSoup(html_title.text, 'lxml')
    for text_itself in html_title:
        text_itself = html_title.find(id='maincontent').text
        texts.append(text_itself)
texts = f(texts)       
```

дата публикации

```python
dates = []
for date1 in hrefs:
    html_title = requests.get(date1)
    html_title.text
    html_title = BeautifulSoup(html_title.text, 'lxml')
    for date_itself in html_title:
        date_itself = html_title.find('div', class_="dcr-km9fgb").text
        date = re.sub(r'GMTLast modified on \w* \d* \w* \d* \d*\.\d* \w*|BSTLast modified on \w* \d* \w* \d* \d*\.\d* \w*', '', date_itself)
        dates.append(date[:-1])
dates = f(dates)       
```

сохранить в датафрейм с колонками "источник", "дата", "заголовок", "текст"

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Источник": hrefs,
                   "Дата": dates,
                   "Заголовок": titles,
                   "Текст": texts})
```
