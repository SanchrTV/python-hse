**Задача 1.**  Напишите генератор названий компаний (почти как генератор паролей). 
Пусть в названии компании будет 6 частей, рандомно выбирающихся из списка names.

Сделайте это функцией

```python
# генератор названий компаний

import random

names = ["пром","агро","торг","урал","север","юг","техно",
"экспо","метал","нефть","сельхоз","фарм","строй",
"кредит","алмаз","-девелопмент","развитие","мос",
"рос","кубань","сибирь","восток","нано","софт",
"микро","онлайн","инвест","текстиль","цемент"]

def unique__capitalize_name(names):
  name = random.sample(names, 6)  #выбираем 6 имен из заданного списка
  str_name = ' '.join(name) #преобразуем список в строку
  name = str_name.title().replace(' ', '')  #выделяем большими буквами и убираем пробелы
  return name

unique__capitalize_name(names)
```

**Задача 2.** Генератор паролей

Пароли, получаемые в результате, должны удовлетворять следующим условиям: 
     - в пароле есть 3 заглавные буквы (любые, в любом месте пароля)
     - в пароле есть 4 цифры (любые, в любом месте)
     - оставшиеся символы пароля - строчные латинские буквы
     
Создайте функцию, которая генерирует пароли, пользователь задает число символов в пароле (как аргумент функции)

```python
import random
import string

password = int(input('Введите желаемое число символов в пароле: '))

def generate_pw(pwd):
  digits = [1, 2, 3, 4, 5, 6, 7, 8, 9]  #чтобы вдальнейшем использовать random.choices
  pwd = []  #создаем лист
  pwd.append(random.choices(digits,k=4))  #в лист вставляем 4 цифры
  pwd.append(random.choices((string.ascii_uppercase),k=3))  #в лист вставляем 3 заглавные буквы
  pwd.append(random.choices((string.ascii_lowercase),k=password-7))   #в лист вставляем оставшие строчные буквы
  pwd = str(pwd) #преобразовать в строку чтобы убрать лишние знаки
  pwd = pwd.replace('[', '')
  pwd = pwd.replace(']', '')
  pwd = pwd.replace(' ', '')
  pwd = pwd.replace(',', '')
  pwd = pwd.replace("'", '')
  return ''.join(random.sample(pwd, len(pwd)))

print('Ваш пароль:', generate_pw(password))
```
