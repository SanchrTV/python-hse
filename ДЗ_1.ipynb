```python
# Задание 1: (условия if-elif-else)

a = float(input('по шкале от 0 до 10, напишите, насколько Вы устали: '))
if a < 0.7:
  print('Кажется, Вы совсем не устали')
elif 0.7 <= a <= 4.8:
  print('Вы немного устали')
elif 4.9 <= a <= 7.9:
  print('Вы достаточно сильно устали')
elif 8 <= a <= 10:
  print('Вам срочно нужен отдых')     
elif a > 10:
  print('ОТДЫХ! ОТДЫХ! И НИЧЕГО КРОМЕ ОТДЫХА!')
```

```python
# Задание 2: (списки и циклы)

import random

c1 = ["мартини","грейпфрутовый сок","жасмин","тоник","лосось"]
c2 = ["клубника","какао","мята","марсала"]
c3 = ["водка","томатный сок","лимонный сок","вустерширский соус","черный перец","сельдерей","лосось"]
c4 = ["джин","вермут","ликер мараскино","апельсины","коктейльная вишня","лосось"]
c5 = ["ром","авокадо","сахарный сироп","сливки","лимонный сок","лед"]
c6 = ["красный вермут","тоник","апельсины","лосось"]
c7 = ["только чай"]

cocktails = [c1, c2, c3, c4, c5, c6, c7]

for i in c1:
  if i == "лосось":
    c1 = c1.remove("лосось")
for i in c2:
  if i == "лосось":
    c2 = c2.remove("лосось")    
for i in c3:
  if i == "лосось":
    c3 = c3.remove("лосось")
for i in c4:
  if i == "лосось":
    c4 = c4.remove("лосось")
for i in c5:
  if i == "лосось":
    c5 = c5.remove("лосось")
for i in c6:
  if i == "лосось":
    c6 = c6.remove("лосось")
for i in c7:
  if i == "лосось":
    c7 = c7.remove("лосось")
print(random.choice(cocktails))           
```
